```python
# Function to calculate sentiment
def calculate_sentiment(text):
    return TextBlob(text).sentiment.polarity

# Applying the function to your dataframe
df['sentiment'] = df['full_text'].apply(calculate_sentiment)
```

```python
df['user_mentions'] = df['entities'].apply(lambda x: [mention['id'] for mention in x['user_mentions']] if 'user_mentions' in x else [])
df['user_mentions_name'] = df['entities'].apply(lambda x: [mention['screen_name'] for mention in x['user_mentions']] if 'user_mentions' in x else [])

df['user_followers_count'] = df['user'].apply(lambda x: x['followers_count'])
df['user_id'] = df['user'].apply(lambda x: x['id'])
df['user_name'] = df['user'].apply(lambda x: x['screen_name'])
```

```python
# Define thresholds
high_follower_threshold = df['user_followers_count'].quantile(0.90)
low_follower_threshold = df['user_followers_count'].quantile(0.25)
```

```python
# Classify tweets
df['category'] = pd.cut(
    df['user_followers_count'],
    bins=[0, low_follower_threshold, high_follower_threshold, float('inf')],
    labels=['Low Follower Count', 'Medium Follower Count', 'High Follower Count']
)
```

```python
df['sentiment_type'] = pd.cut(
    df['sentiment'],
    bins=[-float('inf'), 0, float('inf')],
    labels=['Negative', 'Positive']
)
```

```python
## Change granularity from tweets to mentions
df = df.explode('user_mentions').reset_index(drop=True)

# Explode user_mentions_name similarly
df = df.explode('user_mentions_name').reset_index(drop=True)
```

```python
# Filter subsets
high_follower_positive = df[(df['category'] == 'High Follower Count') & (df['sentiment_type'] == 'Positive')]
high_follower_negative = df[(df['category'] == 'High Follower Count') & (df['sentiment_type'] == 'Negative')]
low_follower_positive = df[(df['category'] == 'Low Follower Count') & (df['sentiment_type'] == 'Positive')]
low_follower_negative = df[(df['category'] == 'Low Follower Count') & (df['sentiment_type'] == 'Negative')]
```

## Mention graph

```python
# Example usage with the high follower count positive segment
```

```python
high_follower_positive.shape
```

```python
# Define the brand names (assuming these are directly mentioned by name in user_mentions_name)
brands = ['lululemon', 'nike', 'adidas']

# Filter the DataFrame to include only rows where the mentioned name is one of the brands
#data_sample = high_follower_positive[high_follower_positive['user_mentions_name'].str.lower().isin(brands)]
data_sample = high_follower_positive.drop_duplicates(subset=['id', 'user_mentions_name', 'user_name']).copy()

min_followers = min(data_sample['user_followers_count'])
max_followers = max(data_sample['user_followers_count'])

data_sample['user_followers_count_norm'] = data_sample['user_followers_count'].apply(lambda x: (x - min_followers) / (max_followers - min_followers))
```

```python
data_sample.shape
```

```python
## Reduce size of data to improve look and feel of the graph
data_sample = data_sample[:1000]
```

```python
G = nx.DiGraph()

# Create nodes and edges with appropriate labels
for _, row in data_sample.iterrows():
    user_id = row['user_id']
    user_name = row['user_name']
    mentioned_id = row['user_mentions']
    mentioned_name = str(row['user_mentions_name']).lower()  # Normalize to lower case for consistency
    followers_count = row['user_followers_count']

    # Add user node with follower count
    if not G.has_node(user_name):
        G.add_node(user_name, followers=followers_count, brand=(user_name in brands))
    
    # Add mentioned user node, assume no follower count data for mentioned names unless specified
    if not G.has_node(mentioned_name):
        is_brand = mentioned_name in brands
        G.add_node(mentioned_name, followers=0, brand=is_brand)

    # Add or update edge with weight
    if G.has_edge(user_name, mentioned_name):
        G[user_name][mentioned_name]['weight'] += 1
    else:
        G.add_edge(user_name, mentioned_name, weight=1)
```

```python
import heapq
from networkx import degree_centrality

n = 25  # Top n central nodes
centrality = degree_centrality(G)
top_central_nodes = heapq.nlargest(n, centrality, key=centrality.get)

# Identify bridgers (simple approximation)
betweenness = nx.betweenness_centrality(G)
top_bridgers = heapq.nlargest(n, betweenness, key=betweenness.get)
```

```python
def visualize_with_pyvis(G):
    net = Network(notebook=True, height="750px", width="100%", bgcolor="#ffffff", font_color="black", directed=True)
    
    for node, attrs in G.nodes(data=True):
        size = attrs['followers'] / 2000  # Scale size by follower count
        color = "#f0a30a" if attrs['brand'] else "#2db7f5"
        if node in top_central_nodes:
            color = "#ff5733"  # Central nodes in red
        if node in top_bridgers:
            color = "#000000"  # Bridgers in black
        net.add_node(node, size=size, color=color, title=f"Followers: {attrs['followers']}")
    
    for source, target, attrs in G.edges(data=True):
        weight = attrs.get('weight', 1)
        title = f"Mentions: {weight}"
        net.add_edge(source, target, value=weight, title=title, arrowStrikethrough=True, color="#0092cc")
        
    # Set physics options to prevent overlap and improve layout
    net.set_options("""
    {
      "physics": {
        "barnesHut": {
          "gravitationalConstant": -30000,
          "centralGravity": 0.3,
          "springLength": 200,
          "springConstant": 0.04,
          "damping": 0.09,
          "avoidOverlap": 0.5
        },
        "solver": "barnesHut",
        "timestep": 0.5
      },
      "nodes": {
        "font": {
          "size": 16,
          "face": "arial"
        },
        "scaling": {
          "label": {
            "enabled": true
          }
        }
      }
    }
    """)

    return net.show("network.html")
```

```python
visualize_with_pyvis(G)
```

## Semantic Network Graph 

```python
## High follower count, positive mentions, Nike

tweets = high_follower_positive.drop_duplicates(subset=['id', 'user_mentions_name', 'user_name']).copy()
tweets = high_follower_positive[high_follower_positive['user_mentions_name'].str.lower().isin(['nike'])]
tweets = tweets.full_text
```

```python
tweets
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

brand_keywords = {
    'nike': ['nike', 'air max', 'uptempo'],
    'adidas': ['adidas', 'ultra boost'],
    'lululemon': ['lululemon']
}

def preprocess_and_tag(text):
    # Convert to lowercase and remove URLs, special characters
    text = text.lower()
    text = re.sub(r'https?\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in word_tokenize(text) if word not in stop_words ]
    #tokens = [tok for tok in tokens if len(tok) > 1]
    return tokens # Generic tag for non-branded words
```

```python
co_occurrences = defaultdict(int)
for tweet in tweets[:1000]:
    words = set(preprocess_and_tag(tweet))  # Ensure uniqueness within the tweet
    for word1, word2 in combinations(words, 2):
        if word1 != word2:
            ordered_pair = tuple(sorted([word1, word2]))  # Sort to ensure consistency
            co_occurrences[ordered_pair] += 1
```

```python
co_occurrences
```

```python
# Build the graph from co-occurrences
G = nx.Graph()
for words, count in co_occurrences.items():
    if count >= 50:
        word1, word2 = words
        G.add_edge(word1, word2, weight=count)
```

```python
len(G)
```

```python
def visualize_graph(G):
    net = Network(notebook=True, height="800px", width="100%")
    for node in G.nodes:
        net.add_node(node, title=str(node))
    for node1, node2, data in G.edges(data=True):
        # Check if brand data exists, then use it
        title = f"{data['weight']} co-occurrences"
        net.add_edge(node1, node2, title=title, value=data['weight'])
        
        
    net.set_options("""
    {
      "physics": {
        "barnesHut": {
          "gravitationalConstant": -30000,
          "centralGravity": 0.3,
          "springLength": 200,
          "springConstant": 0.04,
          "damping": 0.09,
          "avoidOverlap": 0.5
        },
        "solver": "barnesHut",
        "timestep": 0.5
      },
      "nodes": {
        "font": {
          "size": 16,
          "face": "arial",
          "bold": true
        },
        "scaling": {
          "label": {
            "enabled": true
          }
        }
      }
    }
    """)
    return net.show("semantic_network.html")

visualize_graph(G)
```
